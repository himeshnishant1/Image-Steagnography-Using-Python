```python
# Decimal to Binary convertor
def dec2bin(dec):
    inverse = ''
    binary = ''
    while dec != 0:
        inverse = inverse + str( int(dec % 2) )
        dec = int(dec / 2)
    if len(inverse) <= 8:
        size = len(inverse)
        for i in range(8-size):
            inverse = inverse + '0'
    for i in range(len(inverse)-1,-1,-1):
        binary = binary + str(inverse[i])
    return binary
```

```python
# Binary to Decimal convertor
def bin2dec(binar):
    binar = int(binar)
    dec = 0
    i = 0
    while binar > 0:
        dec = dec + (int(binar%10)*np.power(2,i))
        i = i+1
        binar = int(binar/10)
    return dec
```

```python
# XOR operation
def XOR(data,key):
    data = str(data)
    #print("data ",data)
    key = str(key)
    #print("key ",key)
    enc_data = ''
    for ext_data in range(len(data)):
        if data[ext_data] == '0' and key[ext_data] == '0':
            enc_data = enc_data + '0'
        elif data[ext_data] == '0' and key[ext_data] == '1':
            enc_data = enc_data + '1'
        elif data[ext_data] == '1' and key[ext_data] == '0':
            enc_data = enc_data + '1'
        elif data[ext_data] == '1' and key[ext_data] == '1':
            enc_data = enc_data + '0'
    return enc_data
```

```python
def binary_planes_split(img1,m,n):    
    b1=np.zeros((m,n,1),np.uint8)
    b2=np.zeros((m,n,1),np.uint8)
    b3=np.zeros((m,n,1),np.uint8)
    b4=np.zeros((m,n,1),np.uint8)
    b5=np.zeros((m,n,1),np.uint8)
    b6=np.zeros((m,n,1),np.uint8)
    b7=np.zeros((m,n,1),np.uint8)
    b8=np.zeros((m,n,1),np.uint8)

    for i in range(m):
        for j in range(n):
            intent = img1[i,j]
            intent = str(dec2bin(intent))

            
            b1[i,j] = int(intent[7])
            b2[i,j] = int(intent[6])
            b3[i,j] = int(intent[5])
            b4[i,j] = int(intent[4])
            b5[i,j] = int(intent[3])
            b6[i,j] = int(intent[2])
            b7[i,j] = int(intent[1])
            b8[i,j] = int(intent[0])
    return b1,b2,b3,b4,b5,b6,b7,b8
```

```python
def conjugate(mean,std,b_ex):  
    wc = np.zeros((8,8,1),np.uint8)
    bc = np.zeros((8,8,1),np.uint8)
    for i in range(8):
        for j in range(8):
            if int(j % 2) == 0:
                wc[i,j] = 1
                bc[i,j] = 0
            else:
                bc[i,j] = 1
                wc[i,j] = 0
    b_ex_tmp = XOR1(wc,b_ex)
    alpha = alpha1(b_ex_tmp)/63
    if alpha <= (mean-limit*std):
        print("value can not be coded")
        return b_ex
    print("conjugated")
    return b_ex_tmp
```
